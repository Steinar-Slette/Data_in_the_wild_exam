```python
# Load the data
df = pd.read_csv("normalized_output_2022_w_date.csv", sep=';')

# Convert the 'Date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Set 'Date' as the index
df.set_index('date', inplace=True)

# Calculate the total popularity for each keyword
total_popularity = df.sum()

# Select the top 10 keywords
top_10_keywords = total_popularity.nlargest(15)

# Select the data for the top 10 keywords
df_top_10 = df[top_10_keywords.index]

# Plot the popularity of the top 10 keywords over time
df_top_10.plot()
plt.title('Popularity of Top 10 Keywords Over Time')
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.legend(title='Keywords', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
