```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
df_2022 = pd.read_csv('data/processed/articles/2022_articles_processed.csv', sep=";")
df_2023 = pd.read_csv('data/processed/articles/2023_articles_processed.csv', sep=";")

# Count the frequency of each category
category_counts_2022 = df_2022['Category'].value_counts()
category_counts_2023 = df_2023['Category'].value_counts()

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Plot a bar chart of the 2022 category counts
bars_2022 = axs[0].bar(category_counts_2022.index, category_counts_2022.values)
axs[0].set_title('2022 Articles')
axs[0].set_xlabel('Category')
axs[0].set_ylabel('Frequency')
axs[0].tick_params(axis='x', rotation=90)

# Add the total number of articles per category on top of each bar
for bar in bars_2022:
    yval = bar.get_height()
    axs[0].text(bar.get_x() + bar.get_width()/2, yval + 0.5, yval, ha='center', va='bottom')

# Plot a bar chart of the 2023 category counts
bars_2023 = axs[1].bar(category_counts_2023.index, category_counts_2023.values)
axs[1].set_title('2023 Articles')
axs[1].set_xlabel('Category')
axs[1].set_ylabel('Frequency')
axs[1].tick_params(axis='x', rotation=90)

# Add the total number of articles per category on top of each bar
for bar in bars_2023:
    yval = bar.get_height()
    axs[1].text(bar.get_x() + bar.get_width()/2, yval + 0.5, yval, ha='center', va='bottom')

# Adjust the layout
plt.tight_layout()
plt.show()
```
