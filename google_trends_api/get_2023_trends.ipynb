### Get all the tags for which we have pytrends data, along with time series relative popularity. This is to preserve the truly raw data before normalization

```python
import os
import pandas as pd

# Specify the directory where your CSV files are located
directory = r"C:\School\Semester_1\Data_Wrangling\Data_in_the_wild_exam\google_trends_api\raw_results_2023"

# List all CSV files in the directory
csv_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.csv')]

# Initialize an empty DataFrame
request_tags = pd.DataFrame()

# Concatenate all CSV files along the column axis
for file in csv_files:
    df = pd.read_csv(file, sep=';')
    request_tags = pd.concat([request_tags, df], axis=1)

# Specify the output CSV file path
output_csv_path = r"C:\School\Semester_1\Data_Wrangling\Data_in_the_wild_exam\data\intermediate\concated_results_2023\all_tags_2023.csv"

# Write the merged DataFrame to a CSV file with semicolon (;) as the delimiter
request_tags.to_csv(output_csv_path, sep=';', index=False)

# Display a message indicating successful writing
print(f'Merged data has been written to {output_csv_path}')
```

### Get all the UNIQUE tags from the pytrends data, so that we can check that we didn't miss any 

We must create a different version of the all_tags_2022.csv file, in which we don't have the duplicates. 

```python
import os
import pandas as pd

pytrends_unique_tags = []
# Specify the directory where your CSV files are located
directory = r"C:\School\Semester_1\Data_Wrangling\Data_in_the_wild_exam\google_trends_api\raw_results_2023"

# List all CSV files in the directory
csv_files = [os.path.join(directory, file) for file in os.listdir(directory) if file.endswith('.csv')]

# Initialize an empty DataFrame
request_tags = pd.DataFrame()

# Concatenate all CSV files along the column axis
for file in csv_files:
    df = pd.read_csv(file, sep=';')
    request_tags = pd.concat([request_tags, df], axis=1)

request_tags = request_tags.loc[:,~request_tags.columns.duplicated()]
# Specify the path to your CSV file
output_csv_path =  r"C:\School\Semester_1\Data_Wrangling\Data_in_the_wild_exam\data\intermediate\concated_results_2023\all_unique_tags_2023.csv"

# Write the merged DataFrame to a CSV file with semicolon (;) as the delimiter
request_tags.to_csv(output_csv_path, sep=';', index=False)

# Read the CSV file into a DataFrame
df = pd.read_csv(output_csv_path, sep=';')

for column in df.columns:
    # Check if the column is not already in the array
    if column not in pytrends_unique_tags:
        # Add the column to the array
        pytrends_unique_tags.append(column)

# Print the unique columns
print(pytrends_unique_tags)
print(len(pytrends_unique_tags))
```
