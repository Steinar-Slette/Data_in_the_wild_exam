```python
import pandas as pd
```

```python
# Load the data
df_steinar = pd.read_csv("2022_not_complete_sample_annotated_steinar.csv", sep=';')
df_manuel = pd.read_csv("2022_not_complete_sample_annotated_manuel.csv", sep=';')



# Delete rows with 'source' as 'bbc'
df_steinar = df_steinar.loc[df_steinar['Source'] != 'bbc']
df_manuel = df_manuel.loc[df_manuel['Source'] != 'bbc']

# Convert the 'Category' column to lower case
df_steinar['Category'] = df_steinar['Category'].str.lower()
df_manuel['Category'] = df_manuel['Category'].str.lower()

# Ensure the data is in the same order for comparison
df_steinar.sort_values(by='Reference ID', inplace=True)
df_manuel.sort_values(by='Reference ID', inplace=True)

# Compare the 'Category' column in both DataFrames
matches = df_steinar['Category'] == df_manuel['Category']

# Calculate the accuracy
accuracy = matches.mean()

print(f"Accuracy: {accuracy * 100:.2f}%")

# Create a DataFrame with mismatching values
df_mismatch = df_steinar[df_steinar['Category'] != df_manuel['Category']]

# Add the 'Category' column from Manuel's DataFrame to the mismatch DataFrame
df_mismatch['Category_Manuel'] = df_manuel['Category']

# Print the mismatching values
df_mismatch.to_csv(f'mismatches.csv', sep=';')


# Create a DataFrame with matching values
df_match = df_steinar[df_steinar['Category'] == df_manuel['Category']]

# Save the matching values to a CSV file
df_match.to_csv('matches.csv', index=False, sep=';')
```

```python
# Load the data
df_all = pd.read_csv("../data/annotation/_final/2022_January_all_articles_annotated.csv", sep=';')
df_match = pd.read_csv("matches.csv", sep=';')
print(df_match.shape)

# Select the rows in df_all where the 'URL' is in df_match
df_selected = df_all[df_all['Url'].isin(df_match['Url'])]

# Save the selected rows to a CSV file
print(df_selected.shape)
df_selected.to_csv("selected_rows_from_chatgpt.csv", index=False, sep=';')
```

```python
# Load the data
df_selected = pd.read_csv("selected_rows_from_chatgpt.csv", sep=';')
df_match = pd.read_csv("matches.csv", sep=';')

# Merge the dataframes on 'URL'
df_merged = pd.merge(df_selected, df_match, on='Url')

# Save the merged DataFrame to a CSV file
df_merged.to_csv("merged.csv", index=False, sep=';')
```

```python
# Load the data
df_selected = pd.read_csv("selected_rows_from_chatgpt.csv", sep=';')
df_match = pd.read_csv("merged.csv", sep=';')
print(df_selected.shape)
print(df_match.shape)

# Convert '&' to 'and' in the 'Category' column
df_selected['Chat GPT'] = df_selected['Chat GPT'].str.replace('&', 'and')
df_match['Category'] = df_match['Category'].str.replace('&', 'and')

# Ensure the data is in the same order for comparison
df_selected.sort_values(by='Url', inplace=True)
df_match.sort_values(by='Url', inplace=True)

# Reset the index
df_selected.reset_index(drop=True, inplace=True)
df_match.reset_index(drop=True, inplace=True)

# Compare the 'Category' column in both DataFrames
matches = df_selected['Chat GPT'] == df_match['Category']

# Calculate the accuracy
accuracy = matches.mean()

print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Create a DataFrame with mismatching values
df_mismatch = df_selected[df_selected['Chat GPT'] != df_match['Category']]

# Add the 'Category' column from df_match to the mismatch DataFrame
df_mismatch['Category_Match'] = df_match['Category']

# Save the mismatching values to a CSV file
df_mismatch.to_csv('mismatches_chatgpt.csv', index=False, sep=';')
```
