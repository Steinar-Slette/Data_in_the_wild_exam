```python
import pandas as pd
```

```python
# Load the data
df_steinar = pd.read_csv("2023_combined_sample_annotated_v2.csv", sep=';')
df_manuel = pd.read_csv("2023_combined_sample_annotated.csv", sep=';')

# Delete rows with 'source' as 'bbc'
df_steinar = df_steinar.loc[df_steinar['Source'] != 'bbc']
df_manuel = df_manuel.loc[df_manuel['Source'] != 'bbc']

# Convert the 'Category' column to lower case
df_steinar['Category'] = df_steinar['Category'].str.lower()
df_manuel['Category'] = df_manuel['Category'].str.lower()

# Merge the dataframes on 'Reference ID'
df_merged = pd.merge(df_steinar, df_manuel, on='Reference ID', suffixes=('_steinar', '_manuel'))

# Compare the 'Category' column in both DataFrames
matches = df_merged['Category_steinar'] == df_merged['Category_manuel']

# Calculate the accuracy
accuracy = matches.mean()

print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Create a DataFrame with matching values
df_match = df_merged[df_merged['Category_steinar'] == df_merged['Category_manuel']]

# Save the matching values to a CSV file
df_match.to_csv('matches.csv', index=False, sep=';')
```

```python
# Load the data
df_all_articles = pd.read_csv("../../data/annotation/_final/2023_January_all_articles_annotated.csv", sep=';')
df_matches = pd.read_csv("matches.csv", sep=';')

# Merge the dataframes on 'url__manuel'
df_matched = pd.merge(df_all_articles, df_matches, left_on='Url', right_on='Url_manuel', how='inner')

# Save the merged DataFrame to a CSV file
df_matched.to_csv('merged.csv', index=False, sep=';')
```

```python
# Load the data
df_merged = pd.read_csv("merged.csv", sep=';')

# Replace '&' with 'and' in the 'Chat GPT' and 'Category_manuel' columns
df_merged['Chat GPT'] = df_merged['Chat GPT'].str.replace('&', 'and')
df_merged['Category_manuel'] = df_merged['Category_manuel'].str.replace('&', 'and')

# Compare the 'Chat GPT' and 'Category_manuel' columns
matches = df_merged['Chat GPT'] == df_merged['Category_manuel']

# Calculate the accuracy
accuracy = matches.mean()

print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Create a DataFrame with mismatching values
df_mismatch = df_merged[df_merged['Chat GPT'] != df_merged['Category_manuel']]

# Save the mismatching values to a CSV file
df_mismatch.to_csv('mismatches.csv', index=False, sep=';')
```
