```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
# Load the data
df = pd.read_csv("data/", sep=';')

# Count the frequency of each category
category_counts = df['Category'].value_counts()

# Plot the frequencies
category_counts.plot(kind='bar')
plt.title('Frequency of Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

```python
# Load the data
df = pd.read_csv("../data/annotation/01_final_filtered/2022_January_skynews_annotated_cleaned_filtered.csv", sep=';')

# Assuming the keywords are in a column named 'Keywords' and are separated by commas
keywords = df['Tags'].str.split(',').explode()

# Count the frequency of each keyword
keyword_counts = keywords.value_counts()

# Select the top 10 keywords
top_10_keywords = keyword_counts.nlargest(10)

# Plot the frequencies
top_10_keywords.plot(kind='bar')
plt.title('Frequency of Top 10 Keywords')
plt.xlabel('Keyword')
plt.ylabel('Frequency')
plt.show()
```

```python
# Load the data
df = pd.read_csv("../data/annotation/01_final_filtered/2022_January_skynews_annotated_cleaned_filtered.csv", sep=';')

# Convert the string representation of the list to an actual list
df['Tags'] = df['Tags'].apply(ast.literal_eval)

# Separate the keywords into individual rows
df = df.explode('Tags')

# Count the frequency of each keyword
keyword_counts = df['Tags'].value_counts()

# Select the top 10 keywords
top_10_keywords = keyword_counts.nlargest(20)

# Plot the frequencies
top_10_keywords.plot(kind='bar')
plt.title('Frequency of Top 10 Keywords')
plt.xlabel('Keyword')
plt.ylabel('Frequency')
plt.show()
```
